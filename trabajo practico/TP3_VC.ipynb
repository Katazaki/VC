## 1. Obtener una detección del logo en cada imagen sin falsos positivos

###Imagen **coca_logo_2.png**

```python
#leemos imagen 
template = cv2.imread("pattern.png")
template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
template = np.invert(template)
template = cv2.Canny(template, 50, 200)
(tH, tW) = template.shape[:2]
#plt.imshow(template, cmap= 'gray')
```

```python
image = cv2.imread("coca_logo_2.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB )
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#plt.imshow(image, cmap= 'gray')
```

```python
found = None

for scale in np.linspace(1.8, 2, 20)[::-1]:
		# resize the image according to the scale, and keep track
		# of the ratio of the resizing
  resized = imutils.resize(gray, width= int (gray.shape[1]*scale))
  r = gray.shape[1] / float(resized.shape[1])
  if resized.shape[0] < tH or resized.shape[1] < tW:
			break
for angle in np.arange(0,360,15):
  rotated = imutils.rotate(resized, angle)
  w, h = rotated.shape[:: -1]
  
  edged = cv2.Canny(rotated, 50, 200)  
  res = cv2.matchTemplate(edged, template, cv2.TM_CCOEFF_NORMED)
  (_, maxVal, _, maxLoc) = cv2.minMaxLoc(res)


  if found is None or maxVal > found[0]:
      found = (maxVal, maxLoc, r)
  
(_, maxLoc, r) = found
(startX, startY) = (int(maxLoc[0] * r), int(maxLoc[1] * r))
(endX, endY) = (int((maxLoc[0] + tW) * r), int((maxLoc[1] + tH) * r))

cv2.rectangle(image, (startX, startY), (endX, endY), (0, 255,255), 2)

plt.imshow(image)
```

##2. Plantear y validar un algoritmo para múltiples detecciones en la imagen coca_multi.png con el mismo témplate del item 1

```python
#leemos imagen 
template = cv2.imread("pattern.png")
plt.imshow(template)
```

```python
image = cv2.imread("coca_multi.png")
plt.imshow(image)
```

```python
def template_matching(image, template):
  "image = imagen de entrada en la cual se realizara la busqueda despues de ser leida"
  "template = imagen de busqueda, ingresarla despues de ser leida"

  #tratamientos para imagen de entrada
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB )
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  #tratamientos para template
  template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
  template = np.invert(template)
  template = cv2.Canny(template, 100, 200)
  kernel = np.ones((3,3),np.uint8)
  template = cv2.dilate(template,kernel,iterations = 1)
  (tH, tW) = template.shape[:2]

  found = None
  
  for scale in np.linspace(4.1, 5, 20)[::-1]:
		# resize the image according to the scale, and keep track
		# of the ratio of the resizing
      resized = imutils.resize(gray, width= int (gray.shape[1]*scale))
      r = gray.shape[1] / float(resized.shape[1])
      if resized.shape[0] < tH or resized.shape[1] < tW:
        break

      edged = cv2.Canny(resized, 100, 200)
      dilation = cv2.dilate(edged,kernel,iterations = 1)  
      res = cv2.matchTemplate(dilation, template, cv2.TM_CCOEFF_NORMED)
      (_, maxVal, _, maxLoc) = cv2.minMaxLoc(res)

      threshold = 0.5

      loc = np.where( res >= threshold)

      if found is None or maxVal > found[0]:
         found = (maxVal, maxLoc, r)

      if loc: 
         (_, maxLoc, r) = found
         (startX, startY) = (int(maxLoc[0] * r), int(maxLoc[1] * r))
         (endX, endY) = (int((maxLoc[0] + tW) * r), int((maxLoc[1] + tH) * r))
         cv2.rectangle(image, (startX, startY), (endX, endY), (0, 255, 255), 2)

  plt.imshow(image)
```

```python
template_matching(image, template)
```
